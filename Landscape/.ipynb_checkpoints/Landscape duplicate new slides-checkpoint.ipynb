```python
def cleanSectorSegemnt(dfDic):
    catDic,totalDic = {},{}
    for key,df in dfDic.items():
        df=df.copy()
        catIndex = df.index[df['Unnamed: 0'].isin(['Sector','Segment','SubSegment','SubCategory'])][0]
        catName = df['Unnamed: 0'][catIndex]
        df.columns = (df.iloc[catIndex].replace(np.nan,'') + ' '+df.iloc[catIndex-1].ffill().replace(np.nan, "")).str.replace(' Values','').str.strip()#.str.replace('Total ','')
        df = df.iloc[catIndex+1:]
        df[catName] = df[catName].ffill()
        
        df = df.applymap(lambda x: float(x) if pd.to_numeric(x, errors='coerce') == x else x)
        df = df.applymap(lambda x: 0 if pd.isna(x) else x)
#         print(key,df.columns)
        addative = df[~df[catName].str.contains(' Total')].sort_values('WoB %', ascending = False)
        
        total = df[df[catName].str.contains(' Total')]
        total[catName] = total[catName].str.replace(' Total','')

        catDic[key.replace('  ',' ')] = addative
        totalDic[key.replace('  ',' ')] = total
        
    return catDic,totalDic



```

## Slide 13 : Share And Growth

```python
def marketSplit(retailersDic,retailerList,channelList,custList):
    market_regions,market_channels,market_markets = {},{},{}
#     channels_regions,channels_channels,channels_markets = {},{},{}   
#     cust_regions,cust_channels,cust_markets             = {},{},{}        
    total_area = {}
    
    for key,value in retailersDic.items():
        
        table = value.copy()
        table.columns = table.iloc[12]
        table = table.iloc[table[table[table.columns[0]]==table.columns[0]].index[0]+1:]
        table=table.drop(columns=['WoB %', "Relative Price", "Growth Contribution"])
        table.columns = table.columns.str.replace('Trade ', '').str.replace('Channel ','').str.strip()
        table['Value Share'] = table['Value Share'].astype(float)
        table = table.sort_values('WoB %', ascending=False)
        gran_tot = table[table['Region'] == "Grand Total"]


        table['Channel'] = np.where(table['Channel'].isna(), table['Region'], table['Channel'])
        table['Market'] = np.where(table['Market'].isna(), table['Channel'], table['Market'])
        table['Region'] = table['Region'].ffill()

        regionDf = table[(table['Region'].str.contains('Total'))]
        regionDf = regionDf[regionDf.Region.str.replace('Total','').str.strip().isin(retailerList)].drop(columns = ['Channel', 'Market'])
        regionDf['Region'] = regionDf['Region'].str.replace('Total', '').str.strip()

        regionDf = regionDf.drop_duplicates()

        channelDf = table[table['Channel'].str.replace('Total', '').str.strip().isin(channelList)].drop(columns = ['Region','Market'])
        channelDf['Channel'] = channelDf['Channel'].str.replace(' Total', '')

        channelDf = channelDf.drop_duplicates()

        marketDf = table[table['Channel'].isin(custList)].drop(columns = ['Region','Channel'])

        marketDf = marketDf.drop_duplicates()

        if not regionDf.empty:
            market_regions[key] = pd.concat([gran_tot,regionDf])
        if not channelDf.empty:   
            market_channels[key] =  pd.concat([gran_tot,channelDf])
        if not marketDf.empty:
            market_markets[key] =  pd.concat([gran_tot,marketDf])
        if not gran_tot.empty:    
            total_area[key] = gran_tot
    return market_regions,market_channels,market_markets,total_area

    
```

```python
def genrateIndexList(scopeCategory ,chartIndex =0 ,chartCount =4 ):
    if len(scopeCategory) == 0:
        return []
    page = len(scopeCategory) # number of sector or segment or category
 
    lis=[]
    while page >chartCount:
        lis.append(chartCount + chartIndex)
        page -=chartCount
    lis.append(page + chartIndex)
    return [lis]

sectorIndex = genrateIndexList(sectors,4,3)
segmentIndex = genrateIndexList(segments,4,3)
```
