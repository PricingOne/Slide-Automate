```python
def genrateIndexList(scopeCategory, chartIndex=0, chartCount=4):
    """
    Generates a list of chart indices based on the scope category and chart parameters.

    Parameters:
    scopeCategory (list): A list representing the scope categories (e.g., sectors, segments, categories).
    chartIndex (int, optional): The starting index for the first chart. Default is 0.
    chartCount (int, optional): The number of charts to display on a full slide. Default is 4.

    Returns:
    list: A list containing the calculated indices for chart positioning.
    """
    
    # Return an empty list if the scopeCategory is empty
    if len(scopeCategory) == 0:
        return []
    
    # Calculate the number of pages needed
    page = len(scopeCategory)
    
    # Initialize the list to store chart indices
    lis = []
    
    # Loop to calculate indices until pages are exhausted
    while page > chartCount:
        lis.append(chartCount + chartIndex)
        page -= chartCount
    
    # Append the remaining charts to the list
    lis.append(page + chartIndex)
    
    return [lis]
```

```python
def marketcleaning(input_data, catinput, national=False):
    split_dfs = {}
    cat_dfs = {}

    for k in input_data.keys():
        dfbrand = input_data[k].copy()
        dfcat = catinput[k].copy()
        top_brand_col = f"{BrandOrTopB}"

        if national:
            brand_group_col = dfbrand.columns[2]
            cat_group_col = dfcat.columns[1]
            dfbrand = dfbrand[dfbrand[brand_group_col].str.strip().str.upper().isin(['NATIONAL'])]
            dfcat = dfcat[dfcat[cat_group_col].str.strip().str.upper().isin(['NATIONAL', 'NATIONAL TOTAL'])]

        else:
            possible_children = ["Region", "Channel", "Market"]
            existing_children = [col for col in possible_children if col in dfbrand.columns]
            if not existing_children:
                raise ValueError("None of 'Region', 'Channel', or 'Market' columns found in dfbrand")

            for col in possible_children:
                if col in existing_children:
                    child = col
                    break

            brand_group_col = child
            cat_group_col = child

            allowed_values = []
            if child == "Region" and regions is not None:
                allowed_values = [v for val in regions for v in [val, f"{val} total"]]
            elif child == "Channel" and channels is not None:
                allowed_values = [v for val in channels for v in [val, f"{val} total"]]
            elif child == "Market" and markets is not None:
                allowed_values = [v for val in markets for v in [val, f"{val} total"]]

            if allowed_values:
                allowed_values_clean = [val.strip() for val in allowed_values]
                dfbrand = dfbrand[dfbrand[child].str.strip().isin(allowed_values_clean)]
                dfcat = dfcat[dfcat[child].str.strip().isin(allowed_values_clean)]

        # Normalize group keys
        dfbrand['_group_key'] = dfbrand[brand_group_col].str.replace(r'\s*total\s*$', '', case=False, regex=True).str.strip()
        dfcat['_group_key'] = dfcat[cat_group_col].str.replace(r'\s*total\s*$', '', case=False, regex=True).str.strip()

        # Drop rows with missing brand names
        dfbrand = dfbrand.dropna(subset=[top_brand_col])

        # Get unique source keys from dfbrand
        unique_sources = dfbrand['_group_key'].unique()

        for source in unique_sources:
            # Collect all matching rows from full dfbrand (not group_df!)
            brand_source_rows = dfbrand[
                dfbrand[brand_group_col]
                .astype(str)
                .str.strip()
                .str.lower()
                .isin([source.lower(), f"{source.lower()} total"])
            ]

            if brand_source_rows.empty:
                continue

            # Filter cat data as well
            filtered_cat = dfcat[
                dfcat[cat_group_col]
                .astype(str)
                .str.strip()
                .str.lower()
                .isin([source.lower(), f"{source.lower()} total"])
            ]
            cat_dfs[f'{k} | {source}'] = filtered_cat

            for brand in brand_source_rows[top_brand_col].unique():
                if brand in client_brands:
                    split_dfs[f'{k} | {source} | {brand}'] = brand_source_rows[
                        brand_source_rows[top_brand_col] == brand
                    ].copy()

    return split_dfs, cat_dfs
```
